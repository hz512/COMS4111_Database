## Question 4 $-$ Keys

<u>Question</u>

Define the following terms in one or two sentences.
1. Alternate Key
2. Surrogate Key

For each of the above, explain why/when a schema designer would define each type of key.




<u>Answer:</u>
1. An alternate key is a candidate key which is not selected as primary key. Since an alternate key is minimal and unique, we can index alternate keys, which will speed up queries on alternate key considerably.

2. A surrogate key is an attribute that uniquely identifies a row in a table and is not derived from application data. Sometimes it's not easy to find natural primary key or natural primary key is too big. In those cases, using surrogate key saves memory and improves performance. Also, surrogate key does not heavily rely on the stability of source system.

## Question 5 $-$ SQL Concepts

<u>Question</u>

Define the following terms in one or two sentences.
1. (Table) correlation name
2. Lateral clause
3. Scalar sub-query

For each of the above, explain why someone would use the concept in a query.




<u>Answer:</u>
1. An identifier that is used to rename a relation is referred to as a correlation name. It can be used to avoid ambiguity or to establish a correlated reference. A correlation name is also easier to remember and reference. A correlation name is required, in the case of nested table, to identify the result table. 
2. Lateral clause represents a lateral join between two or more tables. It joins the output of the outer query with the output of the underlying lateral subquery. Lateral clause is useful because it allows a subquery in the ```FROM``` clause that is prefixed by the ```LATERAL``` keyword to access attributes of preceding tables or subqueries in the same ```FROM``` clause.
3. A scalar subquery is a subquery that selects only one column or expression and returns one row. A scalar subquery is useful because it can be used wherever an expression returning a value is permitted in an SQL query.

Materialized views allow extremely fast retrieval of aggregate data, as data is precomputed and stored.

2. Cursor
3. Auto-commit.



<u>Answer:</u>

## Question 9 $-$ Query Processing and Optimization

<u>Question</u>

Define the following concepts in a few sentences.
1. Equivalent query
2. Query cost
3. Access paths
4. Conjunctive selection
5. Index, nested loop join
6. Using statistics about data tables to optimize queries.
7. Random sample/sampling

For each of the above, explain the use/benefits.




<u>Answer:</u>
1. Queries that give the same result on any database are equivalent queries. Using equivalent query might optimize the computation, speed up the query and makes coding simpler.
2. Query cost is an estimate based on how many CPU and I/O resources the query optimizer estimates that the query will use. Database server uses query cost estimation to help it choose between execution plans for a particular query.
3. An access path refers to the path chosen by the DBMS to retrieve the requested data after a query is executed. They provide a path through which data can be located and accessed. 
4. A conjunctive selection is a selection of the form <img width=100 src="./Q9_1.png"> Conjunctive selection provides a way of selecting rows that simultaneously satisfy multiple conditions.
5. In indexed nested-loop join, an index is available on the inner loops join attribute, and file scans are replaced with index scans. For each tuple t_r in the outer relation, the index is used to look up tuples in the inner relation  that satisfy the join condition with t_r. Indexed nested-loop join takes advantage of index scan and thus is faster than ordinary nested-loop join.
6. Query optimizer uses statistics on the tables and indexes to determine how many rows a query might need to access for a given query plan. It uses this information to develop its cost estimate for each step in the plan. With the cost estimate, query optimizer can select an execution plan that processes the query most efficiently.
7. Statistics are computed from a sample of the underlying data, instead of examining the entire collection of data. Random sampling refers to the process of selecting the samples randomly. It gives an appropriate representation of the overall underlying data, without excessively emphasizing on any particular part of the relation.

## Question 10 $-$ Transactions

<u>Question</u>

Define the following terms in one or two sentences.
1. Optimistic transaction.
2. Strict two phase locking.
3. Isolation, what levels SQL supports and reasons to choose one of the levels.
4. Deadlock detection, wait-for-graphs
5. ARIES Algorithm.
6. Steal/No-Steal, Force/No-Force
7. Eventual consistency
8. ETag
8. CAP Theorem

For each of the above, explain why/when they apply/their usage/their value.




<u>Answer:</u>
1. Optimistic transaction is a concurrency control scheme applied to transactional systems such as RDBMS. It assumes that trasacrions execute optimistically, meaning that they are frequently able to finish execution and validate at the end without interfering each other. It is generally used in environments with low data contention. In those cases, optimistic transaction is more efficient and leads to higher throughput. 
2. Strict two-phase locking protocol requires not only that locking be two phase, but also that all exclusive-mode locks taken by a transaction be held until that transaction commits. Strict two-phase locking ensures that any data written by an uncommitted transaction are locked in exclusive mode until the transaction commits, preventing any other transaction from reading the data.
3. 
i. __Serializable__: generally ensures serializable execution although exceptions do exist. Serializable is useful for applications that involve sensitive data and require complex queries and updates as it provides consistent data view and effectively runs transactions in serial order.<br>
ii. __Repeatable read__: allows only committed data to be read and requires that, between two reads of a data item by the same transaction, no other transaction is allowed to update it. Repeatable read let us spend minimal effort ensuring that users are getting consistent read view, which means that they are always reading consistent data value within a transaction. <br>
iii. __Read committed__: allows only committed data to be read, but does not require repeatable reads. Read committed is useful when we only want a simple restriction on users from seeing any intermediate, uncommitted, dirty read. <br>
iv. __Read uncommitted__: allows uncommitted data to be read; lowest isolation level allowed by SQL. Read uncommitted 
helps developers debug a transaction by allowing them to see what is happening within the transaction while its still running.
4. __Deadlock__ is a situation where a group of processes/transactions are blocked because each process/transaction is holding a resource and waiting for another resource acquired by some other. A __wait-for graph__ is a graph where the set of vertices consists of all the transactions in the system. And a directed edge from T_i to T_j in the graph implies that transaction Ti is waiting for transaction Tj to release a data item that it needs. If the wait-for graph has a cycle, then there must be a deadlock.
5. ARIES is a recovery algorithm designed to work with a no-force, steal database approach. ARIES uses logs to record the progress of transactions and their actions which cause changes to recoverable data objects. ARIES applies Write-ahead logging, repeats history before the crash in the redo pass, and logs all changes in the undo pass.
6. __steal__: the system is allowed to write modified blocks to disk even if the transactions that made those modifications have not all committed <br>
__no-steal__: blocks modified by a transaction that is still active (hasn't committed) should not be written to disk <br>
__force__: transactions are forced to output all modified blocks to disk when they commit <br> 
__no-force__: a transaction is allowed to commit even if it has modified blocks, which have not yet been written back to disk <br>
7. __Eventual consistency__ is a consistency model which informally guarantees that, if there are no new updates to a given data item, all accesses to that item will eventually retrieve the last updated value. Distributed copies of data can be inconsistent as long as they eventually converge in the absence of further updates. Eventual consistency is preferred when availability is deemed to be more important than consistency.
8. __Entity tage (ETag)__ is an HTTP header used for Web cache validation. ETag allows a client to make conditional requests. ETag is also an identifier for a specific version of a resource. ETag leads to more efficient cache and saves bandwidth, as it eliminates the need to resend a full response if the content has not changed. Also, ETag helps prevent simultaneous updates of a resource from overwriting each other. 
9. __CAP theorem__ states that a distributed system can only simultaneously provide two out of the three desired guarantees -- consistency, availability, and partition tolerance. Thus, in case of a network partition, we have to choose between concistency and availability. In other words, the system can either return an error/time-out if data is not guaranteed to be up-to-date or return the data with no guarantee.
