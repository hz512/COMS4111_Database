## Question 1.1

__Codd's Rule 10: Integrity Independence__

"A database must be independent of the application that uses it. All its integrity constraints can be independently modified without the need of any change in the application. This rule makes a database independent of the front-end application and its interface."

Briefly explain how a relational database management system supports this rule.

_Answer_

Following integrtiy constraints is inherent within RDBMS. First, no attribute in primary key can be NULL. Also, for every distinct foreign key in a referencing table, there must exsit a matching primary key in the source table. Moreover, RDBMS has constraints such as not null and unique, and the system checks these constraints independently.


## Question 1.2

Explain the concepts of _atomic domain_ and _domain constraint._


_Answer_

A domain of an attribute is a set of permitted values for that attribute. A domain is atomic if all elements in the domain are indivisible. A domain constraint of an attribute is a constraint on the set of permitted values on which that attribute can take. For example, not null is a domain constraint.

## Question 1.3

Briefly explain two-tier and three tier applications. What database management system concept connects the application tier and the database tier?


_Answer_

In two tier architecture, application is at the client machine and accesses database system, which is at the server machine, through query language. In three tier architecture, client machine merely serves as front end and does not access database directly. Front end communicates with an application server. The application server communicates with the database system and accesses data. The business logic of the application is embedded in the application tier instead of client machine. 

Application tier access data in the database tier through query language statements.

Why would someone use a procedural language to manipulate data?

__Answer__

DDL: language for specifying the database schema and other properties of the data.
DML: language that enables users to access or manipulate data.
Procedural DML: DML that requires users to specify both what data are wanted and how to get wanted data.
Declarative DML: DML that only requires users to specify what data are wanted.

## Question 1.6

List five responsibilities of a database management system. For each responsibility, explain one problem that would arise if the DBMS did not implement the responsibility.



_Answer_

1. A DBMS is responsible for interacting with file management system. Without this responsibility, a DBMS cannot retrieve content stored in the files. 
2. A DBMS is responsible for enforcing integrity constraints. Without this responsibility, users may enter and get invalid data from DBMS. 
3. A DBMS is responsible for enforcing security. Without this responsibility, users can access unauthorized part of the database. For example, a janitor can access the data which only CEO is allowed to access.
4. A DBMS is responsible for maintaining backup and recovering from failure. Without this responsibility, data may be lost due to unforeseeable events such as electricity outage. 
5. A DBMS is responsible for consistency control. Without this responsibility, data in DBMS may be incorrect. For example, a simultaneous withdrawal and deposit to a bank account will lead to incorrect balance.

## Question 1.8

Briefly explain the concepts of structured data, unstructured data and semi-structured data. Which type of data are documents (Word, HTML, Google Docs) and why?



_Answer_

Structured data is organized into a well-defined data model. Fields of structured data can be pre-designed and are strong typed. All records for a data type has the same field and there are integrity constraints on values. 

Unstructured data is data without explicit records, delimiters, fields or properties. Unstructured data is usually a sequence of characters/bytes that are only interpretable to specific processing engine.

Semi-structured data have some organizational properties and thus can be parsed to extract records and fields into data structures. However, it contains inconsistency and thus cannot be organized into a rigid data model such as relational database. 

Documents are typically semi-structured data because documents share some common consistent and definite components such as header, body, footer. But different documents have different combinations of these components and these components can be arranged and nested in different ways.

## Question 1.10

Explain the concept of and benefits of _physical data independence._




_Answer_

Physical data independence refers to the ability to modify the physical scheme without having to rewrite application programs. In other words, changes in physical level don't propagate to logical or view level. 

Physical data independence allows users to describe a database in conceptual level without having to specify physical structures. Also, with physical data independence, users can easily change physical storage devices without affecting logical scheme.

## Question 2.3

Briefly explain the following terms and provide a very brief example of each:
1. Super key
2. Candidate key
3. Primary key
4. Alternate key (Secondary key)
5. Foreign key
6. Compound key
7. Composite key
8. Surrogate key
9. Natural key
10. External key


_Answer_ <br>
Consider a table Student(ID, first_name, last_name, phone) where ID is the primary key and phone is unique. 
1. Super key is a set of one or more attributes that collectively uniquely identify a row in a table. In the table Family in Question 2.2, family_ID is a super key.
2. A candidate key is a super key for which no proper subset is also a super key. In Family, family_ID is also a candidate key.
3. A primary key is a candidate key chosen by the designer as the principal attribute(s) for uniquely identifying a row in a table. There can be at most one primary key in a table. In Family, family_ID is also a primary key.
4. An alternate key is a candidate key which is not selected as primary key. In Student, phone is an alternate key.
5. A foreign key in a table is one attributes or a combination of attributes that provide and enforce a link between data in two tables. In table FamilyMembers in Question 2.2, family_ID is a foreign key.
7. A composite key is a candidate key which contains two or more attributes. In table FamilyMembers, (first_name, generation, middle_name, family_ID) together forms a composite key.
6. A compound key is a composite key for which each attribute itself is a foreign key. Suppose (column1, column2) in table A is a composite key. Suppose column1 is a foreign key referencing table B and column2 is a foreign key referencing table C, then (column1, column2) is a compound key of table A.
8. A surrogate key is an attribute that uniquely identifies a row in a table and is not derived from application data. Suppose the attribute ID in student is incremental, i.e. student in row 1 has ID 1 and row 2 has ID 2, then ID in Student is a surrogate key
9. A natural key is a candidate key in which every attribute is derived from application data. In Student, phone is a natural key.
10. An exetrnal key is a surrogate key whose domain comes from an external standard. Suppose ID in Student refers to students' SSN, which is defined by the government agency. Then ID is an external key.

## Question 2.5

Explain the following terms:
- Cross-product
- Natural join
- Equi-join
- Theta-join



_Answer_

1. Table C = (table A cross-product table B) means that every row in A is horizontally concatenated with every row in B to form a row in C. If A has m rows and p columns and B has n rows and q columns, then C has mn rows and (p+q) columns. 
2. Natural join is the same as cross-product except row1 from A and row2 from B are horizontally concatenated to form a row in C (= A natural join B) if row1 and row2 have the same values on the attributes that appear in both A and B. Shared attributes between A and B only appear once in C.
3. Equi-join joins tables based on the equivalence of one or more specified attributes (for example, A.column1 = B.column2). The specified attributes don't need to have the same name. Matched (equivalent) columns occur multiple times in the resultant table. Equi-join can be performed on more than two tables.
4. Theta-join joins tables using based on the condition theta. Theta can be any comparison operators. Thus, equi-join is theta-join if theta is an equivalence condition.

CREATE DEFINER=`dbuser`@`localhost` FUNCTION `compute_uni` (first_name VARCHAR(64), last_name VARCHAR(64))

DETERMINISTIC
BEGIN
    declare fn_prefix char(1);
    declare ln_prefix char(1);
    declare uni_pattern varchar(64);
    declare count int;
    set fn_prefix = lower(substr(first_name, 1, 1));
    set ln_prefix = lower(substr(last_name, 1, 1));
    set uni_pattern = concat(fn_prefix, ln_prefix, "%");

RETURN concat(fn_prefix, ln_prefix, count+1);
END$$

CREATE DEFINER=`dbuser`@`localhost` PROCEDURE `create_professor` (
	in first_name varchar(64),
    in last_name varchar(64),
    in email varchar(64),
    in title varchar(64)
)
BEGIN
	declare myuni varchar(64);
    set myuni = compute_uni(first_name, last_name);
    insert into person_t values 
		(myuni, last_name, first_name, email, "p");
	insert into professor_t values
		(myuni, title);
END$$
DELIMITER ;

#### GRANT/REVOKE statements

````python
# grant permissions to standard_user
%sql grant select on w4111midterm.person to "standard_user";
%sql grant select on w4111midterm.professor to "standard_user";
%sql grant select on w4111midterm.student to "standard_user";
%sql grant execute on procedure w4111midterm.create_student to "standard_user";
%sql grant execute on procedure w4111midterm.create_professor to "standard_user";
````

````python
# revoke permissions from standard_user
%sql revoke select on w4111midterm.person from "standard_user";
%sql revoke select on w4111midterm.professor from "standard_user";
%sql revoke select on w4111midterm.student from "standard_user";
%sql revoke execute on procedure w4111midterm.create_student from "standard_user";
%sql revoke execute on procedure w4111midterm.create_professor from "standard_user";
````
